```python
# import all required library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors 
%config InlineBackend.figure_format = 'retina'
```

# Q1 redesign bubble chart

```python
data = {'company': ['Walmart', 'Sinopec Group','State Grid','China National Petroleum','Royal Dutch Shell', 'Saudi Aramco','Volkswagen','BP','Amazon','Toyota Motor'], 
'revenue': [523964, 407009, 383906, 379130, 352106, 329784, 282760, 282616, 280522, 275288], 
'profit': [14881.00, 6793.20, 7970.00, 4443.20, 15842.00, 88210.90, 15542.00, 4026.00, 11588.00, 19096.20]}
```

```python
top10 = pd.DataFrame(data)
top10 = top10.sort_values(by = ['revenue'], ascending=True)
top10
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))


x = np.arange(len(top10.company))

# ax.hlines(y=x, xmin=0, xmax=top10.revenue, color='#666970', lw=4, label='Revenue', alpha=0.5)
revenue = ax.barh(x, top10.revenue, color='#666970', height=0.5, label='Revenue', alpha=0.5 )
profit = ax.barh(x, top10.profit, color='green', height=0.5, label='Profit')

# hide all axis/spines
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.get_xaxis().set_visible(False)

# set x axis cutoff point
ax.set_xlim(0,max(top10.revenue))

# adjust y-axis size and positioning
ax.set_yticks(x)
ax.set_yticklabels(top10.company.sort_values(ascending=True), 
                    fontsize=12,
                    color='#333232',
                     )
yax = ax.get_yaxis()
yax.set_tick_params(color='white')

# create title
plt.rc('text', usetex=False) #turn off latex interpretation
title_start = 5000
inbetween = 60000
gap = 5000
y = 10
ax.text(title_start, y, "Company | ", fontsize='17', color='black', ha ='right', weight='heavy')
ax.text(title_start + inbetween
            , y, "Profit", fontsize='17', color='green', ha ='right', weight='heavy')
ax.text(title_start + inbetween + 3 * gap, y, 'vs', fontsize='17', color='black', ha ='center', weight='heavy')
ax.text(title_start + inbetween + 6.2 * gap, y, "Revenue", fontsize='17', color='#666970', ha ='left', weight='heavy')

# create function to add bar value labels
def autolabel(rects, values, color):
    """
    Attach a text label above each bar displaying its height
    """
    cnt = 0
    for i, rect in enumerate(reversed(rects)):
        width = rect.get_width()
        height = rect.get_height()   
        ax.text(5000 + width, rect.get_y() + 0.145, '$' + f'{round(values[i]/1000, 1)}' +'K', color=color, fontsize=11,
               bbox=dict(facecolor='white', edgecolor='white', pad=0.5, alpha=0.6))

# add bar value labels        
autolabel(revenue, top10.revenue, '#666970')
autolabel(profit, top10.profit, 'green')


# add top margin white space for better border padding
x0, x1, y0, y1 = ax.axis()
ax.axis((x0, x1, y0, y1+1))

plt.tight_layout()
plt.show()


```
